```python
import hashlib

data = []

for hash in hashes:

    # creating an instance for both hash algorithms
    sha256 = hashlib.sha256()
    md5 = hashlib.md5()

    raw_str = hash['raw']

    # concatenating the bytes from each raw string to the sha256 object
    sha256.update(raw_str.encode())
    # storing the hexadecimal digest as a variable
    sha256_str = sha256.hexdigest()
    
    # concatenating the bytes from each raw string to the md5 object
    md5.update(raw_str.encode())
    # storing the hexadecimal digest as a variable
    md5_str = md5.hexdigest()

    data.append({
        "raw": hash['raw'],
        "sha256": "v" if sha256_str == hash['sha256'] else "x",
        "md5": "v" if md5_str == hash['md5'] else "x"
    })
```

```python
# result dataframe
import pandas as pd

def _color_red_or_green(val):
    color = ''

    if val == "x":
        color = 'red'

    if val == 'v':
        color = 'green'

    return 'background-color: %s' % color


df = pd.DataFrame(data)

df1 = df.style.set_table_styles([dict(selector='th', props=[('text-align', 'center')])]).hide(axis='index')
df1.set_properties(**{'text-align': 'center'}).hide_index()
df1.set_properties(subset=['raw'], **{'width': '500px'})
df1.applymap(_color_red_or_green)
```
